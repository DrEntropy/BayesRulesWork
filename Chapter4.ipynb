## Exercise 4.5 (simulate the posterior) and 4.6

For each case, we want to simulate the posteiror by taking draws from the prior and computing the number of days out of n=7 days that the ice cream shop was open.     Then we filter out the cases where k = 3 which was observed.  Finally we want to compute a historgram and the mean.

```python
def simulate_posterior(a, b, k, n, simN):
    ps = beta.rvs(a,b, size =simN)
    cases = zip(ps,binom.rvs(n,ps))
    filtered = [c[0] for c in cases if c[1]==k]
    return filtered
    
```
